#### Analysis of Variance (ANOVA)

```python
# Example: ANOVA between Test Condition and User1's satisfaction
fvalue, pvalue = stats.f_oneway(df_r.groupby('Test Condition (TC)')['User1'].mean())
print(fvalue, pvalue)
```

```python
df_r.groupby('Test Condition (TC)')['User1'].mean()
```

```python
from scipy import stats

# First, get a list of unique test conditions
test_conditions = df_r['Test Condition (TC)'].unique()

# Then, create a list comprehension that extracts User1's satisfaction scores for each test condition
groups = [df_r[df_r['Test Condition (TC)'] == tc]['User1'] for tc in test_conditions]

# Now, use the * operator to unpack the list of groups as arguments to the f_oneway function
fvalue, pvalue = stats.f_oneway(*groups)

print(fvalue, pvalue)
```

```python
from scipy import stats

# Assuming df_r is your DataFrame and it's correctly loaded with the CSV data

# Instead of grouping and taking the mean, directly extract satisfaction scores for User1 for each test condition
# The key here is to ensure each group (i.e., each test condition) has multiple observations

# Create a list of arrays, where each array contains all User1's satisfaction scores for a given test condition
groups = [group["User1"].values for name, group in df_r.groupby('Test Condition (TC)')]

# Now, use the * operator to unpack the list of groups as arguments to the f_oneway function
fvalue, pvalue = stats.f_oneway(*groups)

print(fvalue, pvalue)
```

```python
# Set the size of the plots
plt.figure(figsize=(18, 5))

# Plot for Delay
plt.subplot(1, 3, 1)  # 1 row, 3 columns, 1st subplot
sns.barplot(x='Delay [ms]', y='Mean Satisfaction', data=df_r)
plt.title('Mean User Satisfaction vs. Delay')
plt.xlabel('Delay [ms]')
plt.ylabel('Mean Satisfaction')
plt.ylim(0, 5)  # Set the limits of the y-axis

# Plot for Jitter
plt.subplot(1, 3, 2)  # 1 row, 3 columns, 2nd subplot
sns.barplot(x='Jitter [ms]', y='Mean Satisfaction', data=df_r)
plt.title('Mean User Satisfaction vs. Jitter')
plt.xlabel('Jitter [ms]')
plt.ylabel('')
plt.ylim(0, 5)  # Set the limits of the y-axis

# Plot for Packet Loss Rate
plt.subplot(1, 3, 3)  # 1 row, 3 columns, 3rd subplot
sns.barplot(x='Packet Loss Rate [%]', y='Mean Satisfaction', data=df_r)
plt.title('Mean User Satisfaction vs. Packet Loss Rate')
plt.xlabel('Packet Loss Rate [%]')
plt.ylabel('')
plt.ylim(0, 5)  # Set the limits of the y-axis

plt.tight_layout()  # Adjust the layout to make room for the titles and labels
plt.show()
```

```python
# Define the independent variables (add a constant term to allow the intercept to shift)
X = sm.add_constant(df_r[['Delay [ms]', 'Jitter [ms]', 'Packet Loss Rate [%]']])

# Define the dependent variable
y = df_r['Mean Satisfaction']

# Perform the regression
model = sm.OLS(y, X).fit()

# Print the summary of the regression
print(model.summary())
```

```python
fig = plt.figure(figsize=(18, 5))

# Delay
ax1 = fig.add_subplot(131)
sm.graphics.plot_partregress('Mean Satisfaction', 'Delay [ms]', ['Jitter [ms]', 'Packet Loss Rate [%]'], data=df_r, ax=ax1)

# Jitter
ax2 = fig.add_subplot(132)
sm.graphics.plot_partregress('Mean Satisfaction', 'Jitter [ms]', ['Delay [ms]', 'Packet Loss Rate [%]'], data=df_r, ax=ax2)

# Packet Loss Rate
ax3 = fig.add_subplot(133)
sm.graphics.plot_partregress('Mean Satisfaction', 'Packet Loss Rate [%]', ['Delay [ms]', 'Jitter [ms]'], data=df_r, ax=ax3)

plt.tight_layout()
plt.show()
```
